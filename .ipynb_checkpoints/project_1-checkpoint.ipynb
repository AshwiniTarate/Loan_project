## 2. Data Gathering

```python
test_file = pd.read_csv('loan.csv')
test_file.head()
```

```python
test_file.info()
```

```python
test_file.isna().sum()
```

```python
test_file['emp_length'].value_counts()
```

```python
test_file['delinq_2yrs'].value_counts()
```

```python
test_file['purpose'].value_counts()
```

```python
test_file.drop(['emp_length','delinq_2yrs','addr_state'],axis=1,inplace=True)
```

```python
test_file.head()
```

```python
test_file.isna().sum()
```

```python
test_file['annual_inc'].value_counts()
```

```python
test_file['annual_inc'] = test_file['annual_inc'].fillna(test_file['annual_inc'].median())
```

```python
test_file['revol_util'].value_counts()
```

```python
test_file['revol_util'] = test_file['revol_util'].fillna(test_file['revol_util'].median())
```

```python
test_file['total_acc'].unique()
```

```python
test_file['total_acc'] = test_file['total_acc'].fillna(test_file['total_acc'].median())
```

```python
test_file['longest_credit_length'].unique()
```

```python
test_file['longest_credit_length'] = test_file['longest_credit_length'].fillna(test_file['longest_credit_length'].median())
```

```python
test_file.head()
```

```python
test_file.info()
```

```python
test_file['term'].unique()
```

```python
test_file['months'] = test_file['term'].apply(lambda m : (m.split(' ')[0]))
test_file.months.unique
```

```python
test_file['home_ownership'].value_counts()
```

```python
enc = LabelEncoder()
home_ow_enc = pd.DataFrame(enc.fit_transform(test_file['home_ownership']))
home_ow_enc
```

```python
home_ow_enc.rename(columns={0:"home_own_enc"},inplace=True)
```

```python
home_ow_enc
```

```python
test_file['purpose'].unique()
```

```python
purpose_dummy = pd.get_dummies(test_file['purpose'])
purpose_dummy
```

```python
test_file['verification_status'].unique()
```

```python
test_file['verification_status'].replace({'verified' : 1,'not verified' : 0},inplace = True)
```

```python
file = pd.concat([test_file,home_ow_enc,purpose_dummy],axis=1)
file.head()
```

```python
file.drop(['home_ownership','term','purpose'],axis=1,inplace=True)
```

```python
file.head()
```

```python
file.info()
```

```python
file.months.unique()
```

```python
file['months'] =file['months'].astype('float64')
```

```python
file.isna().sum()
```

```python
file.corr()
```

```python
y = file['bad_loan']
x = file.drop(['bad_loan'],axis='columns')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lo_clf = LogisticRegression()
lo_clf.fit(x_train,y_train)
lo_clf.score(x_test,y_test)
```

#### check wether overfitting or underfitting occurs

```python
y_prediction = lo_clf.predict(x_test) # 40 samples

Test_Dataset_Accuracy = r2_score(y_test,y_prediction)
Test_Dataset_Accuracy
```

```python
y_train_prediction = lo_clf.predict(x_train) # 160 samples

Train_Dataset_Accuracy = r2_score(y_train,y_train_prediction)
Train_Dataset_Accuracy
```
